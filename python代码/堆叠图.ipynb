```python
import numpy as np
import pandas as pd

from bokeh.core.properties import value
from bokeh.plotting import figure, show, output_file, output_notebook
from bokeh.palettes import brewer

import os
os.chdir(r'C:\Users\Kfor\Desktop\深圳杯')

data = pd.read_excel('图数据.xlsx')
#[['年级/科目', '物理', '化学', '生物', '政治', '历史', '地理', '技术']]
data.columns = [i for i in range(7)]
#data.dropna()
data
#output_file('stacked_area.html')

N = 7
df = pd.DataFrame(data).add_prefix('y')
df.index = [str(i) for i in [2017,2018,2019,2020]]
print(df)
p = figure(x_range=[str(i) for i in [2017,2018,2019,2020]], y_range=(0, 2000))
p.grid.minor_grid_line_color = '#eeeeee'

names = ["y%d" % i for i in range(N)]
p.varea_stack(stackers=names,
              x='index',
              color=brewer['Spectral'][N],
              legend=[value(x) for x in [ '物理', '化学', '生物', '政治', '历史', '地理', '技术']], 
              source=df)

# reverse the legend entries to match the stacked order
p.legend[0].items.reverse()

output_notebook()
show(p)
```
